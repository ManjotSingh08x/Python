# Product Analysis

```python
import pandas as pd

data = {
    'Product_ID' : range(101,111),
    'Product_Name': ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Clothing', 'Clothing', 'Food', 'Food', 'Money', 'Money',],
    'Price' : [300, 200, 400, 500, 60,50, 10,15, 25, 20],
    'Stock': [50, 60, 30, 20, 100, 150, 200, 180, 90, 120]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Price'], df['Stock'])
plt.show()
```
